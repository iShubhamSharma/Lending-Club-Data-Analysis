# Data Visualization

```python
plt.figure(figsize=(10,6))
loans[loans['credit.policy']==1]['fico'].hist(bins=15,color = 'blue',label = 'Credit Policy = 1',alpha = 0.6)
loans[loans['credit.policy']==0]['fico'].hist(bins=15,color = 'red',label = 'Credit Policy = 0',alpha = 0.6)
plt.legend()
plt.xlabel('FICO')
```
